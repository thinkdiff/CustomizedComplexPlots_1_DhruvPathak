```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

fig, ax = plt.subplots()

ax.plot(x, y, color='blue', marker='o', label='Prime Growth')

ax.set_title('Prime Numbers Growth')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.legend()
ax.grid(True)

plt.show()
```
